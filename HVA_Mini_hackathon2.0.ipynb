```python
# Function to check if all values in a column have the same data type
def check_column_data_type(df):
    for column in df.columns:
        unique_types = df[column].apply(type).nunique()
        if unique_types > 1:
            print(f"Column '{column}' contains mixed data types.")

check_column_data_type(ev_cat)
check_column_data_type(ev_maker_by_place)
check_column_data_type(operationalPC)
check_column_data_type(vehicle_class)
check_column_data_type(ev_sales_by_makers_and_cat)
```

```python
# Counting number of EV manufacturers by state
ev_makers_by_state = ev_maker_by_place.groupby('State')['EV Maker'].count().reset_index()
ev_makers_by_state.columns = ['State', 'Number of EV Makers']
ev_makers_by_state.head()
```

```python
# Sorting states by number of EV makers and operational charging stations
top_ev_states = state_ev_analysis.sort_values(by=['Number of EV Makers', 'No. of Operational PCS'], ascending=False)
top_ev_states.head()
```
